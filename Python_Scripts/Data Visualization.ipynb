```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
```

```python
def plot_from_csv_agg(file_name, plot_range, bins, x_name, y_name):
    df = pd.read_csv(file_name)
    x_data = df[x_name]
    min_x = min(x_data)
    intervals = np.linspace(min_x, plot_range, bins)
    df = df.groupby(pd.cut(df[x_name], intervals)).sum()
    xArray = np.linspace(min_x, plot_range, bins - 1)
    
    plt.bar(xArray, df[y_name], color ='blue',
            width = 2)
```

```python
def plot_from_csv_non_agg(file_name, x_name, y_name):
    df = pd.read_csv(file_name)
    plt.bar(df[x_name], df[y_name], color ='blue',
            width = 1)
```

```python
plot_from_csv_agg("../IdeaProjects/EV_Orderly_Charging/src/data/dailyTravelDistances.csv", 120, 60, "Distance", "Distribution")
plt.xlabel("Distance(km)")
plt.ylabel("Probability Density")
plt.show()
```
